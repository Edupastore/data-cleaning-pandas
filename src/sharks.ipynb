Observando ahora la columna Date, podemos ver que hay un patrón que se repite, que es 00(día)-Aaa(mes)0000(año). Mediante regex, vamos a indicarle a nuestro dataframe que, todo lo que haya en esa columna y guarde ese patrón, nos elimine el texto y/o carateres que hubiera (si los hubiera) y nos sobreescriba únicamente el patrón, que va a ser la fecha exacta de la observación siguiendo el formato día-mes-año.

```python
sharks['Date'] = sharks['Date'].apply(lambda x: re.sub(r".*(\d{2}-\w{3}-\d{4}).*",r"\1",str(x)) if re.search(r"\d{2}-\w{3}-\d{4}",str(x)) else x)
```

Nos vamos a quedar con estas palabras clave o patrones: swim, surf, fish, bath, diving, snorkel, wading, stand, fell y canoe, ya que parece que son las que más se repiten y son más diferentes entre sí.

_Injury y Fatal (Y/N)_: en Injury vamos a ver si es interesante ver cuáles son los tipos de heridas (o gravedad de éstas) principales y hacer una agrupación similar a la que hicimos en Activity. En Fatal, la respuesta debe ser sí o no y nos hace pensar que esto se refiere a si la persona falleció o no, ya que hemos leído varios archivos PDF (contenidos en las siguientes columnas) y parece que es así la relación. Por lo tanto: Y = muerto y N = no muerto

```python
sharks_cumplen['Injury'].value_counts().head(20) # Esto es un galimatías peor que el de Activity.
```

Primero rellenaremos todos los nulos de ambas columnas con unknown para que nos deje operar después.

En cuanto a la columna Injury, parece que lo que más se repiten son partes del cuerpo y, en concreto, extremidades o partes de las extremidades. El criterio de clasificación que vamos a seguir entonces va a ser por zona dañada, concretamente por extremidad; le vamos a indicar a nuestro dataframe que todo lo que contenga foot, leg, knee, thigh, ankle y calf nos devuelva leg y todo lo que contenga hand, arm, elbow, wrist y shoulder nos devuelva arm.

Previo a esto, vamos a "casar" todos los registros de esta columna que contengan la palabra fatal para que en la columna Fatal sean un Y (yes).

Lo mismo haremos para todo lo que contenga "no injury" en Injury para que en la columna Fatal aparezca N (no).

Hecho esto, todo lo que no sea arm, leg o no injury en la columna Injury, le indicaremos que lo sobreescriba con unknown.

```python
# Rellenamos los nulos reemplazándolos por unknown usando fillna:
sharks_cumplen['Injury'] = sharks_cumplen['Injury'].fillna(value='Unknown')
sharks_cumplen['Fatal (Y/N)'] = sharks_cumplen['Fatal (Y/N)'].fillna(value='Unknown')
```

```python
# Todo lo que contenga "fatal" en Injury debe ser Y en la otra columna:
mask_2 = sharks_cumplen['Injury'].str.contains('fatal', case=False)
```

```python
sharks_cumplen['Fatal (Y/N)'] = np.where(mask_2 & (sharks_cumplen['Fatal (Y/N)'].isnull()), 'Y', sharks_cumplen['Fatal (Y/N)'])
```

```python
# Todo lo que contenga "no injury" en Injury debe ser N en la otra columna:
mask_3 = sharks_cumplen['Injury'].str.contains('no injury', case=False)
```

```python
sharks_cumplen['Fatal (Y/N)'] = np.where(mask_3 & (sharks_cumplen['Fatal (Y/N)'].isnull()), 'N', sharks_cumplen['Fatal (Y/N)'])
```

```python
# Aquí hacemos un filtrado con contain primero y luego le indicamos que reemplace por las palabras que deseemos.
sharks_cumplen.loc[sharks_cumplen['Injury'].str.contains('foot|leg|knee|thigh|ankle|calf', case=False), 'Injury'] = 'leg'
sharks_cumplen.loc[sharks_cumplen['Injury'].str.contains('hand|arm|elbow|wrist|shoulder', case=False), 'Injury'] = 'arm'
```

```python
# Aquí lo mismo, decimos que todo lo que contenga "fatal" nos indique que son más que daños en extremidades y todo lo
# que contenga "no injury" que devuelva "no injury"
sharks_cumplen.loc[sharks_cumplen['Injury'].str.contains('fatal', case=False), 'Injury'] = 'more than limb damage'
sharks_cumplen.loc[sharks_cumplen['Injury'].str.contains('no injury', case=False), 'Injury'] = 'no injury'
```

```python
# Por último, el resto de registros que escapen a estos casos, van a ser catalogados como "unknown".
keywords_2 = ['leg', 'arm', 'more than limb damage', 'no injury']

sharks_cumplen['Injury'] = np.where(~sharks_cumplen['Injury'].isin(keywords_2), 'unknown', sharks_cumplen['Injury'])
```

**5ª Corrección de los tipos de datos: para optimización de memoria.**

```python
# Convertimos las columna date a datetime y Year a integer (no nos deja a datetime)
sharks_cumplen['Date'] = sharks_cumplen['Date'].astype('datetime64[ns]')
```

```python
sharks_cumplen['Year'] = sharks_cumplen['Year'].astype(int)
```

Ahora, vamos a determinar cuáles de entre las 17 filas con Dtype object, pueden ser convertidas a category; vamos a emplear category cuando una columna en nuestro dataframe contenga valores categóricos o de etiqueta que son limitados. Estas columnas van a ser:

Type, Country, Area, Activity, Sex, Injury, Fatal (Y/N) y Species.

```python
columnas_a_cambiar = ['Type', 'Country', 'Area', 'Activity', 'Sex ', 'Injury', 'Fatal (Y/N)', 'Species ' ]

for columna in columnas_a_cambiar:
    sharks_cumplen[columna] = sharks_cumplen[columna].astype('category')
```

```python
# En Time, sustituimos la h por : para cambiar el Dtype a tipo time (aunque se ejecuta el código, no cambia el tipo)
sharks_cumplen['Time'] = sharks_cumplen['Time'].str.replace('h', ':') 
sharks_cumplen['Time'] = pd.to_datetime(sharks_cumplen['Time'], format='%H:%M').dt.time
```

```python
# Por último, vamos a cambiar las columnas restantes a integer tratando de hacer downcast para que disminuya el espacio que
# ocupan en memoria todo lo que se pueda.

for c in sharks_cumplen.select_dtypes('int64'):
    
    sharks_cumplen[c]=pd.to_numeric(sharks_cumplen[c], downcast='integer')
```
